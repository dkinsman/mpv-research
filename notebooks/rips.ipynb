```python
import pandas as pd
import numpy as np
import gudhi as gd
import os
import pickle 
import gudhi.representations
import matplotlib.pyplot as plt
```

```python
with open('../clean/states1.pkl', 'rb') as f:
    states1 = pickle.load(f)
    
with open('../clean/states_multi.pkl', 'rb') as f:
    states_multi = pickle.load(f)

print(len(states1))
print(len(states_multi))
```

```python
#Generate plots for states w/ less than 151 points

directory = '../clean/pd_data'

for files in os.walk(directory):
    if files[0][-2:] in states1:
        #print(files)
        if files[2] == ['lat_long.pkl']:
            #print(files)
            dir = files[0] + r'/lat_long.pkl'
            print(dir)
            state = files[0][-2:]
            with open(dir, 'rb') as f:
                lat_long = pickle.load(f)
        
            stree = gd.RipsComplex(points = lat_long).create_simplex_tree(max_dimension = 2)
            dgm = stree.persistence(persistence_dim_max = True)
            plt.figure()
            gd.plot_persistence_diagram(dgm, legend = True)
            plt.title('Persistance diagram ' + state + ' (rips)')
            plt.savefig('../output/pd/'+ state + r'/rips_pd.jpeg', bbox_inches = 'tight')
        
            stree = gd.AlphaComplex(points = lat_long).create_simplex_tree()
            dgm = stree.persistence(persistence_dim_max = True)
            gd.plot_persistence_diagram(dgm, legend = True)
            plt.title('Persistence diagram ' + state + ' (alpha)')
            plt.savefig('../output/pd/'+ state + r'/alpha_pd.jpeg', bbox_inches = 'tight')
```

```python
rips_complex = gudhi.RipsComplex(points=lat_long,
                                 max_edge_length=24.0)

simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)
result_str = 'Rips complex is of dimension ' + repr(simplex_tree.dimension()) + ' - ' + \
    repr(simplex_tree.num_simplices()) + ' simplices - ' + \
    repr(simplex_tree.num_vertices()) + ' vertices.'
print(result_str)
fmt = '%s -> %.2f'
for filtered_value in simplex_tree.get_filtration():
    print(fmt % tuple(filtered_value))
```
