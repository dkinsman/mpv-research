```python
import numpy as np
import pandas as pd
import pickle
import os
from scipy.spatial import distance_matrix
```

```python
df = pd.read_csv('../raw/mpv-data.csv')
print(df.shape)
df = df[['latitude','longitude', 'state', 'date']]
state = df.state.unique().tolist()
df['date'] = pd.to_datetime(df['date'])
        
#df['date'] = df['date'].dt.strftime('%Y-%m-%d')
print(df.dtypes)
df.head()
```

```python
# CREATES THE FOLDERS for pkls and output images by state

os.chdir('../clean/pd_data')
print(os.getcwd())



for i in state:
    final_directory = os.path.join(os.getcwd(), i)
    #print(final_directory)
    if not os.path.exists(final_directory):
       os.makedirs(final_directory)

os.chdir('../../output/pd')
print(os.getcwd())

for i in state:
    final_directory = os.path.join(os.getcwd(), i)
    #print(final_directory)
    if not os.path.exists(final_directory):
       os.makedirs(final_directory)

os.chdir('../../notebooks')
```

```python
df_states = []
states1=[]
states_multi = []

years = list(range(2013,2023))
print(years)
```

```python
for index, i in enumerate(state):
    print ("State:  ", i)
    df_states.append(df[df.state==i])
    print(df_states[index].shape)
    #df_i.dropna(subset = ['longitude'], axis = 0, how = 'all', inplace = True)
    #df_i.dropna(subset = ['latitude'], axis = 0, how = 'all', inplace = True)
    target_directory = '../clean/pd_data/' + i 
    #print(target_directory)
    
    if df_states[index].shape[0] <= 151:
        target_directory += '/lat_long.pkl'
        df_states[index].drop(['date', 'state'], axis =1, inplace = True)
        lat_long = df_states[index].values.tolist()
        states1.append(i)
        with open(target_directory, "wb") as f:
            pickle.dump(lat_long, f)
    
    else:
        for j in years:
            target_directory = '../clean/pd_data/' + i
            dff = df_states[index][df_states[index].date.dt.year == j]
            lat_long = dff.drop(['date', 'state'], axis =1)
            lat_long = lat_long.values.tolist()
            target_directory += '/lat_long' + str(j) + '.pkl' 
            if j == years[0]:
                states_multi.append(i)
            with open(target_directory, "wb") as f:
                pickle.dump(lat_long, f) 
```

```python

with open('../clean/states1.pkl', "wb") as f:
            pickle.dump(states1, f)
    
with open('../clean/states_multi.pkl', 'wb') as f:
    pickle.dump(states_multi, f)

#df = df[['latitude'], ['longitude']]
#print(df_mi.dtypes)
#print(df_mi.shape)
#df = df.fillna(np.nan).replace([np.nan], [None])
#df_mi.dropna(subset = ['longitude'], axis = 0, how = 'all', inplace = True)
#df_mi.dropna(subset = ['latitude'], axis = 0, how = 'all', inplace = True)


#df.to_csv('../clean/lat_long.csv')
#df_13 = df[(df['date'] > '2013-01-01') & (df['date'] < '2014-01-01')]
#df_13
#df_14 = df[(df['date'] > '2014-01-01') & (df['date'] < '2015-01-01')]
#df_13
#lat_long_mi= df_mi.values.tolist()
```
